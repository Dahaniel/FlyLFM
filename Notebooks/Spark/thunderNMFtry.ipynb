```python
data=NMFmaps.T
```

```python
S=data.shape
```

```python
Demean=data
Dvar=data
Var=np.zeros(S[3])
Tvar=np.zeros(S[3])
Sk=np.zeros(S[3])
Kr=np.zeros(S[3])
```

```python
for i in range(S[3]):
    Demean[:,:,:,i]=data[:,:,:,i]-np.mean(np.mean(np.mean(data[:,:,:,i],0),0),0)
```

```python
for i in range(S[3]):
    Dsq=np.reshape(Demean[:,:,:,i],S[0]*S[1]*S[2])
    Var[i]=np.var(Dsq)
    Dvar[:,:,:,i]=Demean[:,:,:,i]/Var[i]
    Kr[i]=stats.kurtosis(Dsq)
    Sk[i]=stats.skew(Dsq)
```
