```python
ts = data.toSeries()
```

```python
samples = ts.normalize().subset(nsamples=50, thresh=0.05)
```

```python
plt.plot(samples.T);
```

```python
from thunder import KMeans
```

```python
model = KMeans(k=10).fit(ts)
```

```python
from matplotlib.colors import ListedColormap
```

```python
plt.plot(model.centers.T);
```

```python
labels = model.predict(ts)
```

```python
from thunder import Colorize
```

```python
img_labels = labels.pack()
```

```python
newclrs = Colorize.optimize(model.centers, ascmap=True)
```

```python
plt.gca().set_color_cycle(newclrs.colors)
plt.plot(model.centers.T);
```

```python
brainmap = Colorize(totype=newclrs).images(img_labels[:,:,12])
plt.imshow(brainmap);
```

```python
sim = model.similarity(ts)
```

```python
img_sim = sim.pack()
```

```python
plt.imshow(img_sim[:,:,20], cmap='gray',clim=(0,1));
```

```python
brainmap = Colorize(totype=newclrs).images(img_labels[:,:,20], mask=img_sim[:,:,0])
```

```python
plt.imshow(brainmap);
```

```python
brainmap = Colorize(totype=newclrs).images(img_labels, mask=img_sim)
```

```python
from numpy import amax
plt.imshow(amax(brainmap,2));
```
